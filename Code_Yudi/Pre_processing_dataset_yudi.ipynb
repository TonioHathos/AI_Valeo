# Grid Search pour trouver le mieulleure regularization

Essaie avec Multinomial Regression

```python
def MyMultinomialRegression(C=1e-9):
    return make_pipeline(
        StandardScaler(),
        LogisticRegression(multi_class="multinomial", solver="lbfgs", C=C)
    )
```

```python
model_m = MyMultinomialRegression(C=1)
model_m.fit(X_train, y_train)
```

```python
y_pred_m = model_m.predict(X_test)
y_pred_m
```

```python
accuracy_score(y_pred_m, y_test)
```

```python
plot_conf_mat(y_test, y_pred_m, class_names)
```

```python
smote = SMOTE()

X_train_balanced_m, y_train_balanced_m = smote.fit_resample(X_train, y_train)
pd.value_counts(y_train_balanced_m)
```

```python
model.fit(X_train_balanced_m, y_train_balanced_m)
y_pred = model.predict(X_test)
plot_conf_mat(y_test, y_pred,class_names)
```

```python
accuracy_score(y_pred, y_test)
```

```python
print( classification_report(y_test, y_pred) )
```

```python
GridSearchCV(MyMultinomialRegression(), [] ).get_params().keys()
```

```python
k_fold = KFold(n_splits=5, shuffle=True, random_state=5)

parameters_to_test = [{ 'logisticregression__C': [1e-9, 1e-3, 0.01, 0.1, 1, 10, 100, 1000, 10000] }]

# The different scoring metrics available are in:
#     https://scikit-learn.org/stable/modules/model_evaluation.html#scoring-parameter
grid = GridSearchCV(MyMultinomialRegression(), parameters_to_test, cv=k_fold, 
                    scoring='f1_macro')
```

```python
grid.fit(X_train_balanced, y_train_balanced_m)
print("The best regularization weigth is ", grid.best_params_)
```

```python
model = grid.best_estimator_
model.fit(X_train_balanced, y_train_balanced_m)
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
plot_conf_mat(y_test, y_pred, class_names)
```
